```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import stemgraphic
```

```python
price = pd.DataFrame.from_dict ({
    'domestic_price': [4500,3500,6900,8900,1200,3600,4200],
    'foreign_price':  [2500,1500,4900,6900,600,1600,2200]
})
```

```python
plt.title("Domestic Price")
plt.boxplot(price.domestic_price)
plt.show()
```

```python
plt.title("Foreign Price")
plt.boxplot(price.foreign_price)
plt.show()
```

```python
plt.title("Compare of Domestic & Foreign Price")
plt.boxplot(price.domestic_price)
plt.boxplot(price.foreign_price)
plt.xlim(0,10)

plt.show()
```
